```python
from GoogleNews import GoogleNews
import pandas as pd
from time import sleep

googlenews = GoogleNews(start='01/01/2025', end='04/17/2025')
googlenews.search('Tesla')

all_results = []
for page in range(1, 10):
    googlenews.getpage(page)
    results = googlenews.result()
    all_results.extend(results)
    sleep(1) 

df_news = pd.DataFrame(all_results).drop_duplicates(subset=['title', 'link'])
print(df_news.shape)
```
