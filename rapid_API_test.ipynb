```python
import requests
import pandas as pd
from datetime import datetime, timedelta

API_KEY = "87b4335861464062a4e65e9601b89c99"

def get_news(query, from_date, to_date, page=1):
    url = 'https://newsapi.org/v2/everything'

    params = {
        'q': query,
        'from': from_date,
        'to': to_date,
        'language': 'en',
        'sortBy': 'publishedAt',
        'pageSize': 100,   # Max allowed
        'page': page,
        'apiKey': API_KEY
    }

    response = requests.get(url, params=params)
    return response.json()

# Pull Tesla news day-by-day for January 2022 as an example
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 1, 31)

all_articles = []

current_date = start_date
while current_date <= end_date:
    from_str = current_date.strftime('%Y-%m-%d')
    to_str = (current_date + timedelta(days=1)).strftime('%Y-%m-%d')
    print(f"Fetching news from {from_str} to {to_str}")

    page = 1
    while True:
        data = get_news('Tesla', from_str, to_str, page)
        articles = data.get('articles', [])

        if not articles:
            break

        for article in articles:
            all_articles.append({
                'title': article['title'],
                'publishedAt': article['publishedAt'],
                'url': article['url'],
                'description': article.get('description', '')
            })

        if len(articles) < 100:
            break
        else:
            page += 1

    current_date += timedelta(days=1)

# Save to CSV
df = pd.DataFrame(all_articles)
print(df.head())
df.to_csv("tesla_2022_news.csv", index=False)
```
