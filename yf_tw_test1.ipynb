```python
import pandas as pd
import pytz
from datetime import datetime
from alpha_vantage.timeseries import TimeSeries

# === CONFIGURATION ===
api_key = 'IWMJR0CNT9REIC4L'       # Replace with your Alpha Vantage API key
symbol = 'AMZN'                    # Stock ticker
interval = '60min'                 # Granularity
outputsize = 'full'                # 'full' or 'compact'
start_date = '2025-04-01'          # Inclusive start
end_date = '2025-05-05'            # Inclusive end

# === FETCH DATA ===
ts = TimeSeries(key=api_key, output_format='pandas')
data, meta = ts.get_intraday(symbol=symbol, interval=interval, outputsize=outputsize)

# === CLEAN COLUMN NAMES ===
data.rename(columns={
    '1. open': 'Open',
    '2. high': 'High',
    '3. low': 'Low',
    '4. close': 'Close',
    '5. volume': 'Volume'
}, inplace=True)

# === FORMAT DATE ===
data.index = pd.to_datetime(data.index)
data = data.sort_index()
data.index.name = 'Datetime'
data.reset_index(inplace=True)

# === LOCALIZE AND FILTER BY MARKET HOURS (09:30–16:00 ET) ===
eastern = pytz.timezone("US/Eastern")
data['Datetime'] = data['Datetime'].dt.tz_localize('US/Eastern', ambiguous='NaT', nonexistent='NaT').dt.tz_convert('UTC')
data['Datetime_ET'] = data['Datetime'].dt.tz_convert(eastern)

# Keep only rows during US regular market hours (09:30–16:00 ET)
data = data[
    (data['Datetime_ET'].dt.time >= pd.to_datetime("09:30").time()) &
    (data['Datetime_ET'].dt.time <= pd.to_datetime("16:00").time())
]

# === FILTER BY DATE RANGE (in ET) ===
start_dt = eastern.localize(datetime.strptime(start_date, "%Y-%m-%d"))
end_dt = eastern.localize(datetime.strptime(end_date, "%Y-%m-%d") + pd.Timedelta(days=1))

data = data[
    (data['Datetime_ET'] >= start_dt) &
    (data['Datetime_ET'] < end_dt)
]

# === CLEAN FINAL OUTPUT ===
data.drop(columns=['Datetime_ET'], inplace=True)
data[['Open', 'High', 'Low', 'Close', 'Volume']] = data[['Open', 'High', 'Low', 'Close', 'Volume']].astype(float)
tsla_stock_df = data.copy()

# === SAVE / PRINT ===
tsla_stock_df.to_csv(f"{symbol}_filtered_stock_data.csv", index=False)
print(tsla_stock_df.head())
```

```python
import pandas as pd
import pytz
from datetime import datetime
from alpha_vantage.timeseries import TimeSeries

# === CONFIGURATION ===
api_key = 'IWMJR0CNT9REIC4L'       # Replace with your Alpha Vantage API key
symbol = 'AMZN'                    # Stock ticker
interval = '60min'                 # Granularity
outputsize = 'full'                # 'full' or 'compact'
start_date = '2025-04-07'          # Inclusive start
end_date = '2025-04-11'            # Inclusive end

# === FETCH DATA ===
ts = TimeSeries(key=api_key, output_format='pandas')
data, meta = ts.get_intraday(symbol=symbol, interval=interval, outputsize=outputsize)

# === CLEAN COLUMN NAMES ===
data.rename(columns={
    '1. open': 'Open',
    '2. high': 'High',
    '3. low': 'Low',
    '4. close': 'Close',
    '5. volume': 'Volume'
}, inplace=True)

# === FORMAT DATE AND SORT ===
data.index = pd.to_datetime(data.index)
data = data.sort_index()
data.index.name = 'Datetime'
data.reset_index(inplace=True)

# === CONVERT TO US/EASTERN AND FILTER MARKET HOURS ===
eastern = pytz.timezone("US/Eastern")
data['Datetime'] = data['Datetime'].dt.tz_localize('US/Eastern', ambiguous='NaT', nonexistent='NaT')

# Keep only rows between 09:30 and 16:00 ET
data = data[
    (data['Datetime'].dt.time >= pd.to_datetime("09:30").time()) &
    (data['Datetime'].dt.time <= pd.to_datetime("16:00").time())
]

# === FILTER BY DATE RANGE ===
start_dt = eastern.localize(datetime.strptime(start_date, "%Y-%m-%d"))
end_dt = eastern.localize(datetime.strptime(end_date, "%Y-%m-%d") + pd.Timedelta(days=1))

data = data[
    (data['Datetime'] >= start_dt) &
    (data['Datetime'] < end_dt)
]

# === FORMAT DATETIME AS STRING WITH OFFSET ===
data['Datetime'] = data['Datetime'].dt.strftime('%Y-%m-%d %H:%M:%S%z')
data['Datetime'] = data['Datetime'].str.slice(0, -2) + ":" + data['Datetime'].str.slice(-2)  # +HHMM to +HH:MM

# === KEEP ONLY REQUIRED COLUMNS ===
final_data = data[['Datetime', 'Open', 'High', 'Low', 'Close', 'Volume']]

# === SAVE OR PRINT ===
final_data.to_csv(f"{symbol}new_clean_intraday.csv", index=False)
print(final_data.head())
```
